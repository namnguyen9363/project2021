# Task 2: get the info box for all movie

```python
def get_content_value(row_data):
    if row_data.find("li"):
        return [li.get_text(" ",strip=True).replace("\xa0"," ") for li in row_data.find_all('li')]
    elif row_data.find("br"):
        return [text for text in row_data.stripped_strings] #Split long string
    else:
        return row_data.get_text(" ",strip=True).replace("\xa0"," ")
def clean_tags(soup):
    for tag in soup.find_all(['sup','span']):
        tag.decompose()
def get_info_box(url):
    
    r= requests.get(url)
    
    soup = bs(r.content)
        
    info_box = soup.find(class_='infobox vevent')
    info_rows = info_box.find_all("tr")
    
    clean_tags(soup) # remove tag
    
    movie_info = {}
    for index, row in enumerate(info_rows):
        if index == 0:
            movie_info['title']=row.find('th').get_text(" ", strip=True)

        else:
            header = row.find('th')
            if header:
                content_key = row.find('th').get_text(" ", strip=True)
                content_value = get_content_value(row.find('td'))
                movie_info[content_key] = content_value
    return movie_info
```

```python
get_info_box('https://en.wikipedia.org/wiki/One_Little_Indian_(film)')
```

```python
r1= requests.get('https://en.wikipedia.org/wiki/List_of_Walt_Disney_Pictures_films')
soup = bs(r1.content)
content = soup.prettify()
movies = soup.select('.wikitable.sortable i a')

base_path = 'https://en.wikipedia.org/'
movies_info_list = []
for index, movie in enumerate(movies):
    try:   
        title = movie['title']
        
        relative_path = movie['href']
        
        full_patch=base_path +relative_path
        
        movies_info_list.append(get_info_box(full_patch))
    except Exception as e:
        print(movie.get_text())
        print(e)
        
movies_info_list
```

# save/reload movie data

```python
#save
import json
def save_data(title,data):
    with open(title, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=2)
```

```python
#load
def load_data(title):
    with open(title) as f:
        return json.load(f)
```

```python
save_data('disney_data.json',movies_info_list)
```
